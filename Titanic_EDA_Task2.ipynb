```python
# Task 2: Exploratory Data Analysis (EDA) on Titanic Dataset

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Titanic-Dataset.csv")  # Make sure train.csv is uploaded
df.head()

# 1. Generate Summary Statistics
print("\nSummary Statistics:")
print(df.describe())

# 2. Check Nulls & Data Types
print("\nNull Values:")
print(df.isnull().sum())
print("\nData Types:")
print(df.dtypes)

# 3. Histogram for Numerical Columns
df.hist(bins=20, figsize=(12, 8), color='skyblue')
plt.suptitle('Histogram of Numeric Features', fontsize=16)
plt.show()

# 4. Boxplots to Check Outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title('Boxplot of Age and Fare')
plt.show()

# 5. Correlation Matrix
plt.figure(figsize=(10, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# 6. Pairplot (optional)
# sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']])
# plt.show()

# 7. Categorical Analysis - Survival rate by Gender
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival by Gender')
plt.show()

# 8. Categorical Analysis - Survival rate by Class
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival by Class')
plt.show()

# 9. Skewness Check
print("\nSkewness of Numeric Features:")
print(df[['Age', 'Fare']].skew())
```
